**DFT** (J is imaginary unit in python)

```python
def dft_matrix(N):
    # Create the DFT matrix of size NxN
    omega = np.exp(-2j * np.pi / N)  # The primitive root of unity
    n = np.arange(N)
    k = n[:, None]  # Reshape to column vector
    return np.power(omega, k * n)  # Apply the formula


F = dft_matrix(N)

print("DFT Matrix:")
print(F)
```

**Inverse DFT**

```python
def idft_matrix(N):
    # Create the inverse DFT matrix of size NxN
    omega_inv = np.exp(2j * np.pi / N)  # Inverse of the primitive root of unity
    n = np.arange(N)
    k = n[:, None]  # Reshape to column vector
    return np.power(omega_inv, k * n) / N  # Apply the inverse DFT formula
F_inv = idft_matrix(N)

print("Inverse DFT Matrix:")
print(F_inv)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import pandas as pd


N=int(input("Zn= "))





def dft_matrix(N):
    # Create the DFT matrix of size NxN
    omega = np.exp(-2j * np.pi / N)  # The primitive root of unity
    n = np.arange(N)
    k = n[:, None]  # Reshape to column vector
    return np.power(omega, k * n)  # Apply the formula

F = dft_matrix(N)


def idft_matrix(N):
    # Create the inverse DFT matrix of size NxN
    omega_inv = np.exp(2j * np.pi / N)  # Inverse of the primitive root of unity
    n = np.arange(N)
    k = n[:, None]  # Reshape to column vector
    return np.power(omega_inv, k * n) / N  # Apply the inverse DFT formula

F_inv = idft_matrix(N)




IJ=int(input(' Give parameter |I||J|: ' ))


def factor_pairs(m):
    if m <= 0:
        raise ValueError("Only positive integers are allowed.")

    pairs = []
    for i in range(1, int(m**0.5) + 1):
        if m % i == 0:
            pairs.append((i, m // i))

    return pairs


factors=factor_pairs(IJ)  # Output: [(1, 12), (2, 6), (3, 4)]
print(factors)

for factor in factors:
    i=factor[0]
    j=factor[1]
    I=range(0,i)
    J=range(0,j)
    Pi = np.zeros((N, N))
    for i in range(N):
        if i in I:
            Pi[i][i] = 1
    Pj = np.zeros((N, N))
    for i in range(N):
        if i in J:
            Pj[i][i] = 1
    # Initial Bj calculation
    Bj = F_inv @ Pj @ F
    MIJ = Pi @ Bj @ Pi
    initial_eigenvalues = np.sort(np.round(np.linalg.eigvals(MIJ), 2))[::-1]
    x = np.array(range(1,len(initial_eigenvalues)+1))
    plt.plot(x, initial_eigenvalues)
    plt.xlabel('Lambda n')
    plt.ylabel('Value of Lambda')
    plt.title(f'Eigenvalues with I being {len(I)} and J being {len(J)}')
    plt.show()


    print(initial_eigenvalues)
```

```python

import numpy as np
import matplotlib.pyplot as plt

def timelimit(N, M):
    matrix = np.zeros((N, N), dtype=int)
    np.fill_diagonal(matrix, [1 if i < M else 0 for i in range(N)])
    return matrix

def dft_matrix(N):
    omega = np.exp(-2j * np.pi / N)
    n = np.arange(N)
    k = n[:, None]
    return np.power(omega, k * n)

def bandlimit(N, K):
    matrix = np.zeros((N, N), dtype=int)
    np.fill_diagonal(matrix, [1 if i < K or i >= N - K else 0 for i in range(N)])
    return matrix

def idft_matrix(N):
    omega_inv = np.exp(2j * np.pi / N)
    n = np.arange(N)
    k = n[:, None]
    return np.power(omega_inv, k * n) / N

def find_eigenvalues(matrix):
    eigenvalues = np.linalg.eigvals(matrix)
    eigenvalues = np.sort(eigenvalues)[::-1]
    return np.round(eigenvalues, 5)

def plot_eigenvalues(ax, eigenvalue_sets, labels, max_indices, n, m, k):
    """
    Plots multiple sets of eigenvalues in the same subplot.

    Parameters:
    - ax: The subplot axis
    - eigenvalue_sets: List of eigenvalue arrays to plot
    - labels: List of labels corresponding to each eigenvalue set
    - max_indices: Maximum number of indices to plot
    - n, m, k: Parameters for annotation
    """
    colors = ['b', 'r', 'g', 'm', 'c', 'y', 'k']  # Different colors for clarity
    for i, eigenvalues in enumerate(eigenvalue_sets):
        eigenvalues = eigenvalues[:max_indices]  # Limit to first max_indices values
        ax.plot(eigenvalues, marker='o', linestyle='-', color=colors[i % len(colors)], label=labels[i])

    ax.set_title(f'Eigenvalues of Prolate Matrix\n(First {max_indices} Indices)', fontsize=10)
    ax.set_xlabel('Index')
    ax.set_ylabel('Eigenvalue')
    ax.grid(False)
    ax.legend()  # Show legend to differentiate plots

# Create figure and subplots (1 row, 3 columns)
fig, ax = plt.subplots(1, 3, figsize=(18, 4))

for i in range(3):
    eigenvalue_sets = []  # List to store multiple eigenvalue sets
    labels = []  # List to store function labels

    N = int(input('Give value for parameter n: '))
    K = int(input('Give value for parameter k: '))
    M = int(input('Give value for parameter m: '))

    if i == 0:
        N1 = N  # Store max_indices from first input

    W = (2 * K + 1) / N
    time_bandwidth = W * M
    print("Time Bandwidth =", time_bandwidth)

    F = dft_matrix(N)
    F_inv = idft_matrix(N)
    TM = timelimit(N, M)
    TK = bandlimit(N, K)
    BK = F_inv @ TK @ F
    A = TM @ BK @ TM

    # Compute and store eigenvalues for different transformations
    eigenvalues_A = find_eigenvalues(A)
    eigenvalue_sets.append(eigenvalues_A)
    labels.append(f'N = {N} K={K} M={M}')

    N = int(input('Give value for parameter n: '))
    K = int(input('Give value for parameter k: '))
    M = int(input('Give value for parameter m: '))


    W = (2 * K + 1) / N
    time_bandwidth = W * M
    print("Time Bandwidth =", time_bandwidth)

    F = dft_matrix(N)
    F_inv = idft_matrix(N)
    TM = timelimit(N, M)
    TK = bandlimit(N, K)
    BK = F_inv @ TK @ F
    A = TM @ BK @ TM

    eigenvalues_BK = find_eigenvalues(A)
    eigenvalue_sets.append(eigenvalues_BK)
    labels.append(f'N = {N} K={K} M={M}')

    N = int(input('Give value for parameter n: '))
    K = int(input('Give value for parameter k: '))
    M = int(input('Give value for parameter m: '))


    W = (2 * K + 1) / N
    time_bandwidth = W * M
    print("Time Bandwidth =", time_bandwidth)

    F = dft_matrix(N)
    F_inv = idft_matrix(N)
    TM = timelimit(N, M)
    TK = bandlimit(N, K)
    BK = F_inv @ TK @ F
    A = TM @ BK @ TM


    eigenvalues_TM = find_eigenvalues(A)
    eigenvalue_sets.append(eigenvalues_TM)
    labels.append(f'N = {N} K={K} M={M}')

    # Plot multiple eigenvalue sets on the same subplot
    plot_eigenvalues(ax[i], eigenvalue_sets, labels, N1, N, M, K)

plt.tight_layout()  # Adjust layout for better spacing
plt.show()  # Display all subplots
```
